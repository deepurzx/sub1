### 1) Univariate Analysis

```python
#Plotted histogram of Engine size
plt.figure(figsize=(8,6))
plt.hist(data["Engine_Size(L)"],color="purple")
plt.xticks()
plt.yticks()
plt.xlabel("Engine Size")
plt.ylabel("Number of cars")
plt.show()
```

We can observed that most of the cars engine size is  between 2 to 3 litters

```python
# Change the datatype of Fuel type
from sklearn.preprocessing import LabelEncoder
fuel_type=LabelEncoder()
fuel_type.fit(data["Fuel_Type"])
data["Fuel_Type"]=fuel_type.transform(data["Fuel_Type"])
```

```python
# Draw a count plot of Fuel type
plt.figure(figsize=(8,6)) 
ax=sns.countplot(x = 'Fuel_Type',data = data,palette='plasma')
plt.title("No: of cars vs Fuel type")
langs = ['C', 'C++', 'Java', 'Python']
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center")
```

```python
# City and highway fuel consumption details are shown Analysis using kde plot
plt.figure(figsize=(14,6)) 
plt.subplot(121)
sns.kdeplot(data=data, x="Fuel_Consumption_City(L/100 km)",fill=True)
plt.xlabel("Fuel_Consumption_City(L/100 km)")
plt.tight_layout(pad=0.5)

plt.subplot(122)
sns.kdeplot(data=data, x="Fuel_Consumption_Hwy(L/100 km)",fill=True)
plt.xlabel("Fuel_Consumption_Hwy(L/100 km)")
plt.tight_layout(pad=0.5)
```

Plotted a Kde plot of Fuel Consumption of highway and city .We can observe that fuel consumption of city is high compared to the fuel consumption of highway
